6. Define the fit function (energy versus J_i and J_f). We will use the lmfit module  `ExpressionModel` to define the equation we are fitting to. The model as described in the handout is: $$\Delta E = T(1,J_f) - T(0, J_i)=  {\tilde v}_o + B_e[J_f(J_f+1)- J_i(J_i+1)]  -D_e [J_f^2(J_f+1)^2 - J_i^2(J_i+1)^2] - \alpha_e[(3/2)J_f(J_f+1)-(1/2)J_i(J_i+1)],$$ where ${\tilde v}_o$, $B_e$, $D_e$ and $\alpha_e$ are the parameters to fit. Here's the syntax for defining a fit function and the initial guesses for the parameters:
```
energy = lmfit.ExpressionModel('<equation string>',['J_i','J_f'],nan_policy='omit')
lightparams = lmfit.Parameters()
lightparams.add('v_o', value = <starting guess>, vary = True)
lightparams.add('B_e', value = <starting guess>, vary = True)
lightparams.add('D_e', value = <starting guess>, vary = True)
lightparams.add('alpha_e', value = <starting guess>, vary = True)
```

````python
energy = lmfit.models.ExpressionModel('v_o+B_e*(J_f*(J_f+1)-J_i*(J_i+1))-'
                               'D_e*(J_f**2*(J_f+1)**2-J_i**2*(J_i+1)**2)-'
                               'alpha_e*(3/2*J_f*(J_f+1)-1/2*J_i*(J_i+1))',
                               ['J_i','J_f'],nan_policy='omit')
lightparams = lmfit.Parameters()
lightparams.add('v_o', value = 2900, vary = True)
lightparams.add('B_e', value = 9, vary = True)
lightparams.add('D_e', value = 1, vary = True)
lightparams.add('alpha_e', value = 1, vary = True)
````

7. Run the fit. <span style="color:red;font-weight:bold;">If the residuals are not less then your estimated error in the peak positions, go back and check your data entries for typos and mistakes in J_i and J_f.</span> Syntax for running a fit:
```
lightresult = energy.fit(<df>['E(inv_cm)'],lightparams,J_i=<df>['J_i'],J_f=<df>['J_f'], weights=<df>['E_u(inv_cm)'])
lightresult.plot(xlabel="Point", ylabel="Energy ($cm^-1$)")
lightresult
```
