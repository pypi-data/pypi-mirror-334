```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./../../../results-parameters.csv")
grouped = df.groupby("method")
```

```python
fig, ax = plt.subplots()

ax.set_title("Duration per method")

colors = ["tab:blue", "tab:green", "tab:red"]

for i, (method, new_df) in enumerate(grouped):
    ax.plot(new_df["# params"].values,
            new_df['duration'].values,
            label=method,
            marker='x',
            c=colors[i],
            linestyle='solid')

ax.set_ylabel("duration [s]")
ax.set_xlabel("# params")
ax.loglog()


ax.legend(loc="upper left")

secax = ax.twinx()
secax.set_ylabel('duration / parameter [s]')

for i, (method, new_df) in enumerate(grouped):
    secax.plot(new_df["# params"].values,
            new_df['duration per param'].values,
            label=f"{method} (per param)",
            marker='o',
            c=colors[i],
            linestyle='dashed')
secax.loglog()
secax.legend(loc="upper right")
plt.savefig("./../../../results-parameters.png")
```

```python

df = pd.read_csv("./../../../results-artifacts.csv")
grouped = df.groupby("method")
```

```python

fig, ax = plt.subplots()

ax.set_title("Duration per method")

colors = ["tab:blue", "tab:green"]

for i, (method, new_df) in enumerate(grouped):
    ax.plot(new_df["total size [MB]"].values,
            new_df['duration'].values,
            label=method,
            marker='x',
            c=colors[i],
            linestyle='solid')

ax.set_ylabel("duration [s]")
ax.set_xlabel("file size [MB]")
ax.loglog()


ax.legend()

secax = ax.twinx()
secax.set_ylabel('duration / MB [s]')

for i, (method, new_df) in enumerate(grouped):
    secax.plot(new_df["total size [MB]"].values,
            new_df['duration per MB'].values,
            label=f"{method} (per MB)",
            marker='o',
            c=colors[i],
            linestyle='dashed')
secax.loglog()
plt.savefig("./../../../results-artifacts.png")
```
